```python
# demo_calculator.py

def calculate_discount(price, discount_percent):
    """Calculate final price after discount"""
    return price - (price * discount_percent / 100)

def is_eligible_for_free_shipping(total_amount, user_type):
    """Check if order qualifies for free shipping"""
    if user_type == "premium":
        return True
    if total_amount > 50:
        return True
    return False
```

```python
# Prompt 1: Understanding Code
"Explain what the calculate_discount function does and what could go wrong"
```

Answer:- 
#### What it does: 
- calculate_discount(price, discount_percent) returns price minus price * discount_percent / 100 â€” i.e., applies a percentage discount (expects discount_percent as e.g. 20 for 20%).

#### What could go wrong:
- discount_percent or price not numeric (TypeError/ValueError).
- discount_percent outside expected range (e.g., >100 or negative) producing negative or increased prices.
- rounding/float precision issues for money.
- unintended interpretation if caller passes 0.2 expecting 20% (miscommunication about units).
- None or NaN inputs, extremely large values, or missing validation.

```python
# Prompt 2: Finding Issues
"What bugs or edge cases are missing in is_eligible_for_free_shipping function?"
```

Answer:- 
- Missing input validation: no check that total_amount is numeric (int/float/Decimal) or user_type is a string;passing None or non-numeric will raise or behave unexpectedly.
- Boundary error: uses > 50 (excludes exactly 50); business may expect >= threshold.
- Case/whitespace sensitivity: "Premium", " PREMIUM ", or "premium\n" won't match.
- Negative or bogus totals: negative values or string digits ("100") are not handled/validated.
- Floating-point precision: using floats for money can misclassify near-threshold amounts.
- Hard-coded threshold and rules: threshold is fixed and logic doesn't allow different rules by user_type or promotions.
- Unknown user types: unsupported user_type values are silently treated the same as regular users.
- Missing unit tests and docs: no tests for edge cases (exact threshold, negative, type errors).
